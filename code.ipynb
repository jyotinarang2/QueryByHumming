```python
import math
annotation_to_string = ['S']
#print(annotation_to_string)
#print(annotation_to_string[0])
#print(annotation[4500])
new_array = []
for i in range(len(annotation)):
    #print(annotation[i])
    if(annotation[i]<=0):
        new_array.append(0)
    else:
        new_array.append(annotation[i])
print(len(new_array))
print(new_array)
```

```python
#Convert the array to MIDI
array_to_midi = []
for i in range(len(new_array)):
    if(new_array[i] == 0):
        array_to_midi.append(new_array[i])
    else:
        midi_value = 69 + 12*math.log(new_array[i]/440,2)
        array_to_midi.append(midi_value)
print(array_to_midi)
```

```python
i=1
midi_to_string = []
while(i<len(array_to_midi)):
    if(array_to_midi[i] == 0 or (abs(array_to_midi[i+1]-array_to_midi[i])<1):
        midi_to_string.append('S')    
    elif(array_to_midi[i] < annotation[i-1]):
        annotation_to_string.append('D')
    elif(annotation[i] > annotation[i-1]):
        annotation_to_string.append('U')
    i = i+1
```
